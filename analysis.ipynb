```python
courses = pd.read_csv("Data/Courses.csv")
teachers = pd.read_csv("Data/Teachers.csv")
transactions = pd.read_csv("Data/Transactions.csv")
```

```python
transactions["TransactionDate"] = pd.to_datetime(transactions["TransactionDate"])
```

```python
merged = transactions.merge(courses, on="CourseID", how="left")
merged = merged.merge(teachers, on="TeacherID", how="left")
```

```python
merged.head()
```

```python
merged.columns
```

```python
experience_course = merged.groupby(
    ["TeacherID", "TeacherName", "YearsOfExperience"]
)["CourseRating"].mean().reset_index()

experience_course.head()
```

```python
experience_course[
    ["YearsOfExperience","CourseRating"]
].corr()
```

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=experience_course,
    x="YearsOfExperience",
    y="CourseRating"
)

plt.title("Experience vs Course Rating")

plt.show()
```

```python
teacher_course = merged.groupby(
    ["TeacherID","TeacherName","TeacherRating"]
)["CourseRating"].mean().reset_index()

teacher_course.head()
```

```python
teacher_course[
    ["TeacherRating","CourseRating"]
].corr()
```

```python
plt.figure(figsize=(8,5))

sns.regplot(
    data=teacher_course,
    x="TeacherRating",
    y="CourseRating"
)

plt.title("Teacher Rating vs Course Rating")

plt.show()
```

```python
category_rating = merged.groupby("CourseCategory")["CourseRating"].mean().sort_values(ascending=False)

category_rating
```

```python
plt.figure(figsize=(10,5))

category_rating.plot(kind="bar")

plt.title("Average Course Rating by Category")
plt.xlabel("Course Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)

plt.show()
```

```python
level_rating = merged.groupby("CourseLevel")["CourseRating"].mean().sort_values(ascending=False)

level_rating
```

```python
plt.figure(figsize=(8,5))

level_rating.plot(kind="bar", color="orange")

plt.title("Average Course Rating by Course Level")
plt.xlabel("Course Level")
plt.ylabel("Average Rating")

plt.show()
```

```python
heatmap_data = merged.pivot_table(
    values="CourseRating",
    index="CourseCategory",
    columns="CourseLevel",
    aggfunc="mean"
)

plt.figure(figsize=(10,6))

sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu", fmt=".2f")

plt.title("Average Course Rating by Category and Level")

plt.show()
```

```python
enrollment = merged.groupby(
    ["TeacherName", "TeacherRating"]
).size().reset_index(name="Enrollments")

enrollment.head()
```

```python
enrollment["RatingTier"] = pd.cut(
    enrollment["TeacherRating"],
    bins=[0, 3.5, 4.5, 5],
    labels=["Low", "Medium", "High"]
)

enrollment.head()
```

```python
tier_enrollment = enrollment.groupby("RatingTier")["Enrollments"].mean()

tier_enrollment
```

```python
plt.figure(figsize=(7,5))

tier_enrollment.plot(kind="bar", color="green")

plt.title("Average Enrollments by Instructor Rating Tier")
plt.xlabel("Rating Tier")
plt.ylabel("Average Enrollments")

plt.show()
```

```python
expert_teacher = teachers.groupby("Expertise")["TeacherRating"].mean().sort_values(ascending=False)

expert_teacher
```

```python
plt.figure(figsize=(10,5))

expert_teacher.plot(kind="bar")

plt.title("Average Teacher Rating by Expertise")
plt.xlabel("Expertise")
plt.ylabel("Average Teacher Rating")
plt.xticks(rotation=45)

plt.show()
```

```python
expert_course = merged.groupby("Expertise")["CourseRating"].mean().sort_values(ascending=False)

expert_course
```

```python
plt.figure(figsize=(10,5))

expert_course.plot(kind="bar", color="purple")

plt.title("Average Course Rating by Expertise")
plt.xlabel("Expertise")
plt.ylabel("Average Course Rating")
plt.xticks(rotation=45)

plt.show()
```

```python
print("Average Teacher Rating:", round(teachers["TeacherRating"].mean(), 2))

print("Average Course Rating:", round(courses["CourseRating"].mean(), 2))

print("Rating Consistency Index (Std Dev):", round(teachers["TeacherRating"].std(), 2))

print("Experience Impact Score:", round(
    teachers["YearsOfExperience"].corr(teachers["TeacherRating"]), 2
))

print("Enrollment Influence Ratio:", round(
    enrollment["TeacherRating"].corr(enrollment["Enrollments"]), 2
))
```

```python
plt.figure(figsize=(8,6))

corr = merged[[
    "Age",
    "YearsOfExperience",
    "TeacherRating",
    "CourseRating",
    "CoursePrice",
    "Amount"
]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Matrix")

plt.show()
```

# Key Insights

## Instructor Performance
- The instructor ratings show the overall teaching quality across the EduPro platform.
- Experienced instructors generally demonstrate better teaching performance, although the relationship may vary in strength.
- Some instructors consistently outperform others, indicating opportunities for mentoring and knowledge sharing.

## Course Quality
- Certain course categories receive higher learner ratings than others.
- Course levels influence learner satisfaction, with some levels achieving consistently higher ratings.
- High-quality content contributes significantly to learner satisfaction.

## Revenue & Transactions
- Monthly revenue trends highlight seasonal changes in learner enrollments.
- Payment methods are distributed across multiple options, indicating diverse learner preferences.

## Expertise Analysis
- Some expertise domains consistently achieve higher instructor and course ratings.
- These domains can serve as benchmarks for instructor training and curriculum development.

# Business Recommendations

1. Introduce mentoring programs where highly rated instructors guide newer instructors.

2. Review lower-rated course categories and update course content regularly.

3. Encourage instructors to participate in continuous professional development.

4. Use instructor performance metrics during course assignment and promotion decisions.

5. Increase marketing efforts for high-performing course categories.

6. Monitor instructor ratings regularly through automated performance dashboards.

7. Develop targeted training programs for expertise areas with lower learner satisfaction.

8. Use data-driven KPIs to evaluate instructor performance objectively.

# Conclusion

This project developed a comprehensive data-driven framework to evaluate instructor performance and course quality on the EduPro platform.

The analysis demonstrated how instructor ratings, teaching experience, expertise, and course characteristics influence learner satisfaction and overall platform performance.

By identifying top-performing instructors, evaluating course quality across categories and levels, and measuring the relationship between instructor performance and enrollments, EduPro can make informed decisions to improve educational outcomes.

The findings support continuous improvement through objective performance monitoring, instructor development programs, and evidence-based decision making.
